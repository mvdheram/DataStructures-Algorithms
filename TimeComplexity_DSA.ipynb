```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```
def exponential(n):
  return np.power(n,2)
```

```
def nlogn(n):
  return n*(np.log(n))
```

```
def logn(n):
  return np.log(n)
```

```
def feature_representation(n):
  return [n,exponential(n),nlogn(n),logn(n),1]
```

```
def _instance_features(n):
  evidence_features = pd.DataFrame([feature_representation(i) for i in range(1,n)],columns=  ['n','n2','nlogn','logn','constant'])
  return evidence_features
```

```
features = _instance_features(20)
```

fig - Container that holds the  plot elements \\
ax - Part of the plot that holds data 

```
fig, ax = plt.subplots(figsize=(10,7))
linestyles = ['-', '--', '-.', ':']
markers = ['o' , 'v' , '.', ',']
color =['red','green','orange','blue']
for index,linestyle in enumerate(linestyles):
  ax.plot(features.iloc[:,0],features.iloc[:,index+1],marker= markers[index],linestyle = linestyle,color = color[index],label = features.columns[index+1])
ax.legend()
ax.set_xlabel("N(values)")
ax.set_ylabel("Transfer-function (values) ")
ax.set_title("Time Complexity graph")
```
