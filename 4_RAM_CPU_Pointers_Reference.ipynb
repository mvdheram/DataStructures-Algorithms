Background:

* **RAM (Random Access Memory)**
  * Program instruction and data stored in memory 
    * *High-level program instruction* ------(translated, compiled, executed)------▶ *low level machine instruction* 
    * Done with the help of compilers or interpretors or a combinaiton
      * Differences :
        * Compiler : 
          * **Translation done for the entire program into intermediate code (object code) *saved, and then* executed by CPU.** 
          * Faster execution but slower translation 
        * Interpretor : 
          * **Line-by-line instructions are translated (byte code) and excuted on virtual machine (*no intermediate code saved*).**
          * Slower execution but faster translation.
    * Different types/levels of memory based on the distance from the processor
      * **Registers (closer,smaller, faster) -> Cache -> RAM -> Virtual memory (far, larger, slower)**
      * Types of memroy caches  
        * L1 cache (8KB) ⏭ L2,L3 cache (larger size and slower) ⏭ Main memory (RAM) (larger size and slower)

* **CPU(Processor)** [2]
  * Does the computation of the machine level instruction.
  * Consists of instruction set 
    * E.g. Load, store, jump and logical operations etc. 
  * **Hardware** 
    * **Control Unit**
      * Main unit responsible for controlling the flow of instruction.
      * 3 main stages per instruction (fetch ➡ decode ➡ execute ♻)
        * Fetch 
          * Fetch next instruction from memory
        * Decode 
          * Deterine required action based on instruction set 
        * Execute
          * Execute the instruction 
    * **ALU (Arithmetic Logic Unit)**
      * Responsible for all the arithematic and logical operations(decisions) based on instruction set. 
    * **Registers (general purpose)**
      * Small high speed memory (caches) within processor  used for loading and storing the instruction (arithematic and logical) and data.

* **Data transfer between CPU and RAM**
  * Data transfer occurs through **bus**  
    * Bus -▶ connections (circuit) connecting CPU to memory and other components.
    * Types of bus :
      * Address bus ▶ Used to store the data to RAM address (CPU ➡ RAM)
      * Data bus ▶ Used to send and receive data from RAM (CPU ➡ RAM ♻ )  
  * Unit of bits transfered (bus width/ register width /) is called a **word**
    * **Word** ▶ Natural unit of data used by particular processor[1]. 
    * **Each unique memory location in RAM holds one word of information.**
  * **Word size** ▶ 64 bit or 32 bit CPU processors 

* CPU performance 
  * CPU Cores ▶ CPU containing multiple processing units (ALU,control unit, registers)
  * CPU speed ▶ Number clock cycles per second (number of fetch-execute cycles per second)
    * Measured in hertz (*Hz*)
      * E.g.: 3.2 GHz ▶ 3.2 billion cycles per second
    * Clock cycles ▶ Amount of time between two pulses (CPU clock) of an oscillator. Used to synchronise and coordinate all the components of CPU.
